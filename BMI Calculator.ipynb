```python
# User inputs for weight and height
wt = int ( input ("Enter weight in pounds: ") )
ht = int ( input ("Enter height in inches: ") )

# Below if-else statement: error checking
# If either the height or weight values entered by the user are 0 or less, an error message will be returned.
# Otherwise, the BMI will be calculated

if (wt <= 0) or (ht <= 0):
    print ("ERROR: Invalid input(s). Please enter inputs for weight and height greater than 0.")

else:
    
    # BMI calculation using the height and weight
    BMI = (wt * 703) / ht ** 2
    BMI = round (BMI, 1)
    
    # Calculated BMI output
    print ("Your BMI is:", BMI)

    # Below if-else statement will output the status based on the conditions
    if BMI < 18.5:
        print ("You are underweight.")
    elif BMI <= 24.9:
        print ("You are normal weight.")
    elif BMI <= 29.9:
        print ("You are overweight.")
    elif BMI <= 34.9:
        print ("You are obese.")
    elif BMI <= 39.9:
        print ("You are severly obese.")   
    else:
        print ("You are morbidly obese.")   
```

```python
import pandas as pd

df = pd.DataFrame ({
        'name': ['Nelson', 'Liliam', 'Gabriel', 'Abuelito'],
        'weight': [219, 142, 53, 124],
        'height': [71, 61, 40, 63]
    })
```

```python
print (df)
```

```python
df['BMI'] = df ['weight'] / df['height']**2 * 703
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.describe()
```

```python
def status (df):
    if df ['BMI'] < 18.5:
        return 'Underweight'
    elif (df ['BMI'] >= 18.5) & (df ['BMI'] < 25):
        return 'Normal weight'
    elif (df ['BMI'] >= 25) & (df ['BMI'] < 30):
        return 'Overweight'
    else:
        return 'Obese'
    
df['status'] = df.apply(lambda x: status(x), axis=1)
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
#chart_order = ['Abuelito', 'Gabriel', 'Liliam', 'Nelson']
plt.bar('name','BMI',data=df, color="blue")
plt.xlabel('Name')
plt.ylabel('BMI')
plt.title('BMI of Family Members')
plt.xticks(df['name'])
```

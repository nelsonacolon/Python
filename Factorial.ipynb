```python
def factorial (x):
    product = 1
    for i in range (1, x+1):
        product *= i
    
    return product

print("Enter a number greater than 0.\nThis will compute the factorial of the number entered.")
number = int(input())

if number <= 0:
    print("Sorry, {} is an invalid entry. Please enter a number greater than 0.".format(number))
else:
    result = factorial(number)
    print("{}! = {}".format(number,result))
    
```

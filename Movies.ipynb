```python
# Pandas library
import pandas as pd
```

```python
# Reading data from source file
Movie_data = pd.read_csv('Movie-Data.csv')
```

```python
# Output first 5 rows of dataframe
Movie_data.head()
```

```python
# Statistical information for columns that are numbers
Movie_data.describe()
```

```python
# Output last 5 rows of dataframe
Movie_data.tail()
```

```python
# Outputs data types of all columns in dataframe
Movie_data.dtypes
```

```python
# Release date converted from string to date data type
Movie_data['Release Date'] =  pd.to_datetime(Movie_data['Release Date'])
```

```python
# Removes the '$' and ',' characters from the values in the 'Budget' and 'Revenue' columns
Movie_data['Revenue'] = Movie_data['Revenue'].str.replace('$','')
Movie_data['Revenue'] = Movie_data['Revenue'].str.replace(',','')
Movie_data['Budget'] = Movie_data['Budget'].str.replace('$','')
Movie_data['Budget'] = Movie_data['Budget'].str.replace(',','')

# Converts the data type of the 'Budget' and 'Revenue' columns from string to float
Movie_data['Budget'] = Movie_data['Budget'].astype('float')
Movie_data['Revenue'] = Movie_data['Revenue'].astype('float')
```

```python
# Converts the data type of the 'Budget' and 'Revenue' columns from float to int64
Movie_data['Budget'] = Movie_data['Budget'].astype('int64')
Movie_data['Revenue'] = Movie_data['Revenue'].astype('int64')
```

```python
# Outputs information for each column in dataframe such as count of non-null values and data type
Movie_data.info()
```

```python
# Checks for any duplicate values in dataframe
Movie_data['dupes'] = Movie_data.duplicated()
Movie_data[Movie_data['dupes'] == True].count()
```

```python
# Outputs the number of null values for each column in the dataframe
Movie_data.isnull().sum()
```

```python
# Columns that are not needed for analysis are removed from dataframe
Movie_data = Movie_data.drop(['Wikipedia URL', 'Director (1)', 'Director (2)', 
                              'Cast (1)', 'Cast (2)', 'Cast (3)', 'Cast (4)',
                               'Cast (5)'], 
                              axis=1)
```

```python
# Outputs first 10 records of dataframe
Movie_data.head(10)
```

```python
# An additional column that is not needed is removed from dataframe
Movie_data = Movie_data.drop(['dupes'], axis=1)
```

```python
# New columns are created for month, quarter, year, week and day of week of movie release
Movie_data['Release month'] = Movie_data['Release Date'].dt.month_name()
Movie_data['Release quarter'] = Movie_data['Release Date'].dt.to_period('Q').dt.strftime('%Y-Q%q')
Movie_data['Release year'] = Movie_data['Release Date'].dt.strftime('%Y')
Movie_data['Release week'] =  Movie_data['Release Date'].dt.isocalendar().week
Movie_data['Release weekday'] =  Movie_data['Release Date'].dt.day_name()
```

```python
# column names are converted to snake case
from skimpy import clean_columns

Movie_data = clean_columns(Movie_data)
```

```python
# The values of the budget and revenue columns are divided by 1 million
# This will simplify reporting and visualization
Movie_data['budget'] = Movie_data['budget'] / 1000000
Movie_data['revenue'] = Movie_data['revenue'] / 1000000
```

```python
# Additional libraries for data analysis and visualization
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Scatter plot to show relationship between budget and revenue
plt.scatter(Movie_data['budget'],Movie_data['revenue'])
plt.xlabel("Budget (in millions)")
plt.ylabel("Revenue (in millions)")
plt.title("Movie Budget vs Revenue")
```

```python
# Aggregation for sum of revenue grouped by release_weekday
wkday_sum = Movie_data.groupby(['release_weekday'])['revenue'].sum().reset_index()
wkday_sum.head(12)
```

```python
# Aggregation for sum of revenue grouped by release_month
month_sum = Movie_data.groupby(['release_month'])['revenue'].sum().reset_index()
month_sum.head(12)
```

```python
# release_year converted from string to integer data type
Movie_data['release_year'] = Movie_data['release_year'].astype('int')
```

```python
# Aggregation for sum of revenue grouped by release year
year_sum = Movie_data.groupby(['release_year'])['revenue'].sum().reset_index()
year_sum.head(12)
```

```python
year_sum.head(10)
```

```python
year_sum.tail(10)
```

```python
# Aggregation for sum of revenue grouped by genre
genre_sum = Movie_data.groupby(['genre'])['revenue'].sum().reset_index()
genre_sum.head(10)
```

```python
# Aggregation for count of movies grouped by genre
genre_count = Movie_data.groupby(['genre'])['movie_title'].count().reset_index()
genre_count.head(10)
```

```python
# Bar plot that shows revenue by day of week
plt.figure(figsize=(10,6));

day_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 
               'Saturday']

sns.barplot(
    data = wkday_sum,
    x = 'release_weekday',
    y = 'revenue',
    order = day_order,
    color = "blue",
    errorbar = None);
plt.xlabel("Day of Week");
plt.ylabel("Revenue (in millions)");
plt.title("Movie Revenue by Day of Week");
```

```python
# Bar plot that shows revenue by month
plt.figure(figsize=(10,6));

month_order = ['January', 'February', 'March',
               'April', 'May', 'June',
               'July', 'August', 'September',
               'October', 'November', 'December']               

sns.barplot(
    data = month_sum,
    x = 'release_month',
    y = 'revenue',
    order = month_order,
    color = "blue",
    errorbar = None);
plt.xlabel("Month");
plt.ylabel("Revenue (in millions)");
plt.title("Movie Revenue by Month");
plt.xticks(rotation=45);
```

```python
# Bar plot that shows revenue by year
plt.figure(figsize=(10,6));            

sns.barplot(
    data = year_sum,
    x = 'release_year',
    y = 'revenue',
    color = "blue",
    errorbar = None);
plt.xlabel("Year");
plt.ylabel("Revenue (in millions)");
plt.title("Movie Revenue by Year");
```

```python
# Bar plot that shows revenue by genre
plt.figure(figsize=(10,6));            

sns.barplot(
    data = genre_sum,
    x = 'genre',
    y = 'revenue',
    color = "blue",
    errorbar = None)
plt.xlabel("Genre")
plt.ylabel("Revenue (in millions)")
plt.title("Movie Revenue by Genre")
plt.xticks(rotation=45)
```

```python
# Matplotlib Line plot that shows revenue by year
plt.figure(figsize=(10,6));            

plt.plot(year_sum['release_year'],year_sum['revenue'])
plt.xlabel("Year");
plt.ylabel("Revenue (in millions)")
plt.title("Movie Revenue by Year")
plt.show()
```

```python
# Seaborn Line plot that shows revenue by year
sns.lineplot(data=year_sum, x='release_year', y='revenue')
plt.xlabel("Year");
plt.ylabel("Revenue (in millions)")
plt.title("Movie Revenue by Year")
plt.show()
```

```python
# import urllib library
from urllib.request import urlopen
  
# import json
import json

# store the URL in url as 
# parameter for urlopen
url = "https://data.cityofnewyork.us/resource/erm2-nwe9.json?$query=SELECT%0A%20%20%60unique_key%60%2C%0A%20%20%60created_date%60%2C%0A%20%20%60closed_date%60%2C%0A%20%20%60agency%60%2C%0A%20%20%60agency_name%60%2C%0A%20%20%60complaint_type%60%2C%0A%20%20%60descriptor%60%2C%0A%20%20%60location_type%60%2C%0A%20%20%60incident_zip%60%2C%0A%20%20%60status%60%2C%0A%20%20%60community_board%60%2C%0A%20%20%60borough%60%2C%0A%20%20%60open_data_channel_type%60%2C%0A%20%20%60latitude%60%2C%0A%20%20%60longitude%60%2C%0A%20%20%60%3A%40computed_region_92fq_4b7q%60%2C%0A%20%20%60%3A%40computed_region_sbqj_enih%60%0AWHERE%0A%20%20(%60created_date%60%0A%20%20%20%20%20BETWEEN%20%222022-01-01T00%3A00%3A00%22%20%3A%3A%20floating_timestamp%0A%20%20%20%20%20AND%20%222023-07-05T22%3A22%3A17%22%20%3A%3A%20floating_timestamp)%0A%20%20AND%20((%60closed_date%60%20IS%20NOT%20NULL)%20AND%20caseless_eq(%60status%60%2C%20%22Closed%22))%0AORDER%20BY%20%60created_date%60%20DESC%20NULL%20FIRST"
  
# store the response of URL
response = urlopen(url)
  
# storing the JSON response 
# from url in data
data_json = json.loads(response.read())
```

```python
# Creating a new location type field named location_type2 to have fewer unique values for location type
# Value for location_type2 field will be based on conditions for the location_type field in each row
def new_loc (location_type):
    if (location_type == 'RESIDENTIAL BUILDING') | (location_type == 'Residential Building/House') | (location_type == '3+ Family Apt. Building') | (location_type == '3+ Family Apartment Building') |(location_type == '1-2 Family Dwelling'):
        return 'Residential'
    elif (location_type == 'Street/Sidewalk') | (location_type == 'Street') | (location_type == 'Sidewalk') | (location_type == 'Street/Curbside') | (location_type == 'Gutter') | (location_type == 'Alley'):
        return 'Street/Sidewalk'
    elif (location_type == 'Business') | (location_type == 'Club/Bar/Restaurant') | (location_type == 'Commercial Building') | (location_type == 'Store/Commercial'):
        return 'Business/Commercial'
    elif (location_type == 'Subway'):
        return 'Subway'
    elif (location_type == 'Taxi'):
        return 'Taxi'
    elif (location_type == 'Park/Playground') | (location_type == 'Park'):
        return 'Park'
    else:
        return 'Other/Unknown'

df['location_type2'] = df['location_type'].map(new_loc)

#new_loc = {
#     'Street/Sidewalk': ["Street/Sidewalk", "Street", "Sidewalk", "Gutter", "Alley","Street/Curbside"],
#     'Residential':['RESIDENTIAL BUILDING','Residential Building/House','3+ Family Apt. Building', '3+ Family Apartment Building', '1-2 Family Dwelling'],
#    'Business/Commercial':['Business','Club/Bar/Restaurant','Commercial Building','Store/Commercial'],
 #    'Subway':['Subway'],
 #    'Taxi':['Taxi'],
 #    'Park':['Park/Playground','Park'],
 #    'Other/Unknown':['Other (Explain Below)','Gutter','Alley','Vacant Lot']
#}

#df['location_type2'] = df['location_type'].map(new_loc)
```

```python
def is_prime (number):
    
    counter = 0
    
    for i in range (2, number):
        if number % i == 0:
            counter += 1
            
    if counter == 0:
        return True
    else:
        return False
    

print("Enter a number greater than or equal to 2.")
input_number = int(input())
    
if input_number < 2:
    print ("Invalid entry")
else:    
    if is_prime(input_number) == True:
        print ("{} is a prime number.".format(input_number))
    else:
        print ("{} is a not prime number.".format(input_number))
        
            
```

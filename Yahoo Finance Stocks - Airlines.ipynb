```python
# Obtain stock tickers for the four largest U.S. airlines
aal = yf.Ticker("AAL")   # American Airlines Group Inc
dal = yf.Ticker("DAL")   # Delta Air Lines, Inc.
luv = yf.Ticker("LUV")   # Southwest Airlines Co
ual = yf.Ticker("UAL")   # United Airlines Holdings Inc
```

```python
# Ticker data for the past rolling 10 years for each airline converted into its own dataframe
aal_hist = pd.DataFrame( aal.history(period="10y") )
dal_hist = pd.DataFrame( dal.history(period="10y") )
luv_hist = pd.DataFrame( luv.history(period="10y") )
ual_hist = pd.DataFrame( ual.history(period="10y") )
```

```python
# Dataframe information for one of the companies
luv_hist.info()
```

```python
# Ratio plot for Southwest - comparing the open and close prices
luv_hist[['Open','Close']].iplot(kind='ratio')
```

```python
# Time-series chart comparing the closing prices for the four companies during the past rolling 10 years
plt.figure(figsize=(22,6))
plt.plot(aal_hist['Close'], color='cyan', label='American')
plt.plot(dal_hist['Close'], color='purple', label='Delta')
plt.plot(luv_hist['Close'], color='brown', label='Southwest')
plt.plot(ual_hist['Close'], color='blue', label='United')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.title("4 Largest U.S. Airline Stock Prices - Past 10 Years")
plt.legend()
plt.show()
```

```python
# All 4 dataframes concatenated -- this new dataframe is named 'df'
df = pd.concat([aal_hist,dal_hist,luv_hist,ual_hist], axis=0)
```

```python
# Output first 5 rows of concatenated dataframe
df.head()
```

```python
# Info on df dataframe
df.info()
```

```python
# Reset index numbers
# The date column is currently used the index for the df dataframe
# The data column is converted to a column
# This is important for analysis
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'Date'})
```

```python
# Date column converted to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Info of df dataframe after Date field has been converted from index to column.
df.info()
```

```python
# Output first 5 rows of df dataframe
df.head()
```

```python
# Highest closing prices of each airline in rolling 10-year period
df.groupby('ticker').max()['Close']
```

```python
# Lowest closing prices of each airline in rolling 10-year period
df.groupby('ticker').min()['Close'].sort_values(ascending=False)
```

```python
# Standard deviation of closing prices of each airline in rolling 10-year period. Sort from highest to lowest.
df.groupby('ticker').std()['Close'].sort_values(ascending=False)
```

```python
# 'return' field created to calculate stock return for each day
df['return'] = df.groupby('ticker')['Close'].pct_change()
```

```python
# A new dataframe named returns that will include the returns of each airline during the rolling 10-year period
returns = df[['Date','ticker', 'return']]
```

```python
# A pivot table for returns dataframe
# This will be used for further statistical analysis
returns_pivot = pd.pivot_table(data=returns, values='return', index='Date', columns='ticker', aggfunc='sum')
```

```python
# These dataframes include the means and standared deviations of returns for each of the 4 airlines
# The time period is from Q1 to Q3 2023
returns_mean = pd.DataFrame(returns_pivot.loc['2023-01-01':'2023-09-30'].mean())
returns_std = pd.DataFrame(returns_pivot.loc['2023-01-01':'2023-09-30'].std())
```

```python
# columns renamed in the returns_mean dataframe
returns_mean.reset_index(inplace=True)
returns_mean = returns_mean.rename(columns = {'ticker':'company'})
returns_mean = returns_mean.rename(columns = {0:'mean'})
```

```python
# columns renamed in the returns_std dataframe
returns_std.reset_index(inplace=True)
returns_std = returns_std.rename(columns = {'ticker':'company'})
returns_std = returns_std.rename(columns = {0:'std'})
```

```python
# View the output of the returns_mean dataframe after the renaming of the columns
returns_mean.head()
```

```python
# View the output of the returns_std dataframe after the renaming of the columns
returns_std.head()
```

```python
# The returns_mean and returns_std dataframes joined
# This will create the ret_summary dataframe which will be used for visualization of returns vs. risk
ret_summary = returns_mean.merge(returns_std, how='inner')
```

```python
# Info on ret_summary dataframe
ret_summary.info()
```

```python
# Output of ret_summary dataframe
print(ret_summary)
```

```python
# Scatterplot - stock returns vs. risk analysis
ax = ret_summary.plot(kind='scatter',x='mean', y='std')

ax.set(title='Stock Returns vs. Risk', xlabel='Daily Expected Returns (%)', ylabel='Risk')

# For loop -- labeling points in scatterplot
for idx, row in ret_summary.iterrows():
    ax.annotate(row['company'], (row['mean'],row['std']), size=10, ha='center', va="bottom")
```

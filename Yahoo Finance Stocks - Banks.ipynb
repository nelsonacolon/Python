```python
# This is used to for data ETL using the Yahoo Finance API
import yfinance as yf
```

```python
# Imports for EDA, data cleaning, and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Optional Plotly Method Imports
import plotly
import cufflinks as cf
cf.go_offline()
```

```python
# Obtain stock tickers for the largest U.S. banks
bac = yf.Ticker("BAC")   # Bank of America
c = yf.Ticker("C")       # Citgroup
gs = yf.Ticker("GS")     # Goldman Sachs
jpm = yf.Ticker("JPM")   # JPMorgan Chase
ms = yf.Ticker("MS")     # Morgan Stanley
wfc = yf.Ticker("WFC")   # Wells Fargo
```

```python
# Ticker data for the past rolling 10 years for each bank converted into its own dataframe
bac_hist = pd.DataFrame( bac.history(period="10y") )
c_hist = pd.DataFrame( c.history(period="10y") )
gs_hist = pd.DataFrame( gs.history(period="10y") )
jpm_hist = pd.DataFrame( jpm.history(period="10y") )
ms_hist = pd.DataFrame( ms.history(period="10y") )
wfc_hist = pd.DataFrame( wfc.history(period="10y") )
```

```python
# Dataframe information for Citi
c_hist.info()
```

```python
# Spread plot for Citi - comparing the open and close prices
c_hist[['High','Low']].iplot(kind='spread')
```

```python
# Ratio plot for Citi - comparing the open and close prices
c_hist[['Open','Close']].iplot(kind='ratio')
```

```python
# Time-series chart comparing the closing prices for the 3 largest U.S. banks during the past rolling 10 years
plt.figure(figsize=(22,6))
plt.plot(c_hist['Close'], color='cyan', label='Citi')
plt.plot(jpm_hist['Close'], color='brown', label='JPMorgan Chase')
plt.plot(bac_hist['Close'], color='blue', label='Bank of America')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.title("3 Largest U.S. Bank Stock Prices - Past 10 Years")
plt.legend()
plt.show()
```

```python
# Time-series chart comparing the closing prices for the 3 largest U.S. banks during the past rolling 10 years
# A larger view
plt.figure(figsize=(18,8))
plt.plot(c_hist['Close'], color='cyan', label='Citi')
plt.plot(jpm_hist['Close'], color='brown', label='JPMorgan Chase')
plt.plot(bac_hist['Close'], color='blue', label='Bank of America')
plt.plot(gs_hist['Close'], color='grey', label='Goldman Sachs')
plt.plot(ms_hist['Close'], color='purple', label='Morgan Stanley')
plt.plot(wfc_hist['Close'], color='orange', label='Wells Fargo')
plt.xlabel("Year")
plt.ylabel("Closing Price")
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.title("U.S. Bank Stock Prices - Past 10 Years")
plt.legend()
plt.show()
```

```python
# A column named ticker is added to each dataframe. 
# To be used when concatenating these dataframes.
bac_hist['ticker'] = 'BAC'
c_hist['ticker'] = 'C'
gs_hist['ticker'] = 'GS'
jpm_hist['ticker'] = 'JPM'
ms_hist['ticker'] = 'MS'
wfc_hist['ticker'] = 'WFC'
```

```python
# Output of the first 5 rows of data from the Wells Fargo dataframe
wfc_hist.head()
```

```python
# All 6 dataframes concatenated -- this new dataframe is named 'df'
df = pd.concat([bac_hist,c_hist,gs_hist,jpm_hist,ms_hist,wfc_hist], axis=0)
```

```python
c_hist.columns
```

```python
# Reset index numbers
# The date column is currently used the index for the df dataframe
# The data column is converted to a column
# This is important for analysis
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'Date'})
```

```python
# Date column converted to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Info of df dataframe after Date field has been converted from index to column.
df.info()
```

```python
# Output first 5 rows of df dataframe
df.head()
```

```python
# Highest closing prices of each bank in rolling 10-year period
df.groupby('ticker').max()['Close']
```

```python
# Lowest closing prices of each bank in rolling 10-year period
df.groupby('ticker').min()['Close'].sort_values(ascending=False)
```

```python
# Standard deviation of closing prices of each bank in rolling 10-year period. Sort from highest to lowest.
df.groupby('ticker').std()['Close'].sort_values(ascending=False)
```

```python
# Average (mean) of closing prices of each bank in rolling 10-year period. Sort from highest to lowest.
df.groupby('ticker').mean()['Close'].sort_values(ascending=False)
```

```python
# 'return' field created to calculate stock return for each day
df['return'] = df.groupby('ticker')['Close'].pct_change()
```

```python
# Output first 5 rows of df to verify if return field appears
df.head()
```

```python
# Output last 5 rows of df to verify if return field appears
df.tail()
```

```python
# Output rows with null values in return field. This is the first day of data within rolling 10-year period
df[df['return'].isna()]
```

```python
# A new dataframe named returns that will include the returns of each airline during the rolling 10-year period
returns = df[['Date','ticker', 'return']]
```

```python
# Output first 5 rows of returns dataframe
returns.head()
```

```python
# A pivot table for returns dataframe
# This will be used for further statistical analysis
returns_pivot = pd.pivot_table(data=returns, values='return', index='Date', columns='ticker', aggfunc='sum')
```

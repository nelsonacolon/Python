```python
import yfinance as yf
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
citi = yf.Ticker("C")
```

```python
citi.actions
```

```python
#citi.dividends
```

```python
citi_hist = citi.history(period="1mo")
citi_hist = pd.DataFrame(citi_hist)
```

```python
print(citi_hist)
```

```python
citi_hist.info()
```

```python
#citi_hist.plot(y='Close')
plt.plot(citi_hist['Close'], color='blue', label='Close')
#plt.plot(citi_hist['Open'], color='orange', label='Open')
plt.xlabel("Year")
plt.ylabel("Price")
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.title("Citi Stock Price Per Share: Rolling One Month\nStock Ticker Symbol: C")
#plt.legend()
plt.show()
```

```python
# Pfizer
pfe = yf.Ticker("PFE")
pfe_hist = pfe.history(period="24mo")
pfe_hist = pd.DataFrame(pfe_hist)
```
